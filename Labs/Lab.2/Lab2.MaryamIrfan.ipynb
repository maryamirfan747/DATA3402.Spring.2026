*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
def create_game(n):
    if n <= 0:
        return "Error: n must be a positive integer"

    board = []

    for x in range(n):
        row = []
        for y in range(n):
            row.append(0)
        board.append(row)
    return board
````

*Exercise 2:* Write a function that takes two integers `n` and `m` and **draws** an `n` by `m` game board.

For example, the following is a 3×3 board:

```
   --- --- --- 
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- --- 
```


````python
def draw_game(n, m):
    if n <= 0 or m <= 0:
        return "Error: n and m must be positive integers"

    board = []

    for i in range(n):
        row = []
        for j in range(m):
            row.append(0)
        board.append(row)
    return board
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
def check(board):
    n = len(board)

    for row in board:
        if row.count(row[0]) == n and row[0] != 0:
            return row[0]

    for col in range(n):
        first = board[0][col]
        if first != 0:
            win = True
            for row in range(n):
                if board[row][col] != first:
                    win = False
            if win:
                return first

    first = board[0][0]
    if first != 0:
        win = True
        for i in range(n):
            if board[i][i] != first:
                win = False
        if win:
            return first

    first = board[0][n - 1]
    if first != 0:
        win = True
        for i in range(n):
            if board[i][n - 1 - i] != first:
                win = False
        if win:
            return first

    for row in board:
        if 0 in row:
            return -1 
    return 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
def mark(board, player, row, col):
    n = len(board)

    if row < 0 or row >= n or col < 0 or col >= n:
        return False

    if board[row][col] != 0:
        return False

    board[row][col] = player
    return True
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
def draw_board_labeled(board):
    n = len(board)

    col_labels = "   " + "   ".join(str(i+1) for i in range(n))
    print(col_labels)

    for i, row in enumerate(board):
        row_label = chr(ord('A') + i)
        row_cells = []
        for cell in row:
            if cell == 1:
                row_cells.append("X")
            elif cell == 2:
                row_cells.append("O")
            else:
                row_cells.append(" ")
        print(f"{row_label}  " + " | ".join(row_cells))
        
        if i < n - 1:
            print("   " + "---+" * (n-1) + "---")
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
def mark_labeled(board, player, location):
    
    if len(location) < 2:
        return False
    
    n = len(board)

    row_char = location[0].upper()
    row_index = ord(row_char) - ord('A')

    try:
        col_index = int(location[1:]) - 1
    except:
        return False 

    if row_index < 0 or row_index >= n or col_index < 0 or col_index >= n:
        return False

    if board[row_index][col_index] != 0:
        return False

    board[row_index][col_index] = player
    return True
````

````python
board = [[0, 0, 0],[0, 0, 0],[0, 0, 0]]

print(mark_labeled(board, 1, "B2"))  
print(mark_labeled(board, 2, "B2"))  
print(mark_labeled(board, 2, "C1"))  
print(mark_labeled(board, 1, "D1"))  

draw_board_labeled(board)
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
def player_turn(board, player):

    while True:
        move = input(f"Player {player} ({'X' if player==1 else 'O'}), enter your move (Example: A1, B3): ").strip()
        
        if mark_labeled(board, player, move):
            print(f"Move accepted: {move}")
            break  
        else:
            print("Invalid move. Try again.")
````

````python
def game():
    size = 3  
    board = create_game(size)
    current_player = 1
    
    while True:
        draw_board_labeled(board)
        player_turn(board, current_player)
        status = check(board)

        if status == 1:
            draw_board_labeled(board)
            print("Player 1 (X) wins")
            break
        elif status == 2:
            draw_board_labeled(board)
            print("Player 2 (O) wins")
            break
        elif status == 0:
            draw_board_labeled(board)
            print("The game is a draw")
            break

        current_player = 2 if current_player == 1 else 1
````

````python
def minimax(board, player, depth, max_depth):

    winner = check(board)
    if winner == 2:
        return 10 - depth  
    elif winner == 1:
        return depth - 10  
    elif winner == 0:
        return 0

    if depth >= max_depth:
        return 0  

    n = len(board)
    if player == 2:
        best_score = -1000
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 2
                    score = minimax(board, 1, depth + 1, max_depth)
                    board[i][j] = 0
                    if score > best_score:
                        best_score = score
        return best_score
    else:
        best_score = 1000
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 1
                    score = minimax(board, 2, depth + 1, max_depth)
                    board[i][j] = 0
                    if score < best_score:
                        best_score = score
        return best_score

#Used ChatGPT to help understand how to write the code
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
def computer_move(board, player, max_depth):
    n = len(board)
    best_score = -1000 if player == 2 else 1000
    best_move = None

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = player
                score = minimax(board, 1 if player == 2 else 2, 1, max_depth)
                board[i][j] = 0

                if player == 2 and score > best_score:
                    best_score = score
                    best_move = (i, j)
                elif player == 1 and score < best_score:
                    best_score = score
                    best_move = (i, j)

    if best_move:
        board[best_move[0]][best_move[1]] = player
        
#Used ChatGPT to write majority of the code
````

````python
def com_vs_com(n, depth_p1, depth_p2, first_player):
    board = create_game(n)
    player = first_player

    while True:
        if player == 1:
            computer_move(board, 1, depth_p1)
        else:
            computer_move(board, 2, depth_p2)

        result = check(board)
        if result != -1:
            return result

        player = 2 if player == 1 else 1
````

````python
def run(n, smart_depth, weak_depth):
    smart_wins = 0
    games = 10

    for g in range(games):
        if g % 2 == 0:
            result = com_vs_com(n, weak_depth, smart_depth, 1)
            if result == 2:
                smart_wins += 1
        else:
            result = com_vs_com(n, smart_depth, weak_depth, 1)
            if result == 1:
                smart_wins += 1

    win_rate = smart_wins / games
    print(f"{n}x{n} board: smarter player win rate = {win_rate:.2f}")
    
#Used ChatGPT to help write the code
````

````python
run(3, smart_depth=9, weak_depth=2)
run(4, smart_depth=4, weak_depth=1)
run(5, smart_depth=3, weak_depth=1)
````
